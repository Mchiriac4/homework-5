```python
# Question 8
# Problem Introduction
    # Inspired by the Fisher Tea Experiment, we are seeing if 49/80 STA130 students being able to identify if milk was poured first is  statistically significant or just a radom guess.
# Relationship to Fisher’s Tea Experiment
    # The diffrnce between the Fisher experiment and our own is the population, the Fisher experiment was done on one person, we are testing the general student population, as well as sample, the Fisher experiment was done with 8 cups of tea, we are testing 80 students with one cup each.
# Statements of the Null Hypothesis and Alternative hypothesis
    # The Null Hypothesis assumes that students are guessing if the milk was poured first, making H0: p=0.5, the p value should be the same as a random guess, 50%.
    # The Alternitive Hypothesis suggest that students are able to tell what was poured first, making H1: p>0.5, the p value should be grater than guessing. 
```

```python
# Question 8 

# Quantitative Analysis

# The sample proportion = 49/80 = 0.6125 and the null hypothesis = 0.5 

# Methodology Code and Explanations

import numpy as np
from scipy.stats import binomtest, norm
# Set seed for reproducibility
np.random.seed(42)
# Observed data
n = 80  # Sample size
x = 49  # Number of correct identifications
p_null = 0.5  # Null hypothesis proportion
# Binomial test using binomtest (new function)
p_value = binomtest(x, n, p=p_null, alternative='greater').pvalue
print(f"P-value: {p_value}")
# Normal approximation
p_hat = x / n  # Sample proportion
se = np.sqrt(p_null * (1 - p_null) / n)  # Standard error
z = (p_hat - p_null) / se  # Z-score
p_value_z = 1 - norm.cdf(z)  # One-tailed p-value
print(f"Z-test P-value: {p_value_z}")

# Supporting Visualizations

import matplotlib.pyplot as plt
# Parameters
n_simulations = 10000
null_guesses = np.random.binomial(n, p_null, size=n_simulations)
# Plotting the sampling distribution
plt.figure(figsize=(8, 6))
plt.hist(null_guesses, bins=np.arange(25, 60, 1), density=True, color='skyblue', edgecolor='black')
plt.axvline(x=x, color='red', linestyle='--', label=f'Observed Correct Guesses (x = {x})')
plt.title('Sampling Distribution of Correct Guesses under Null Hypothesis (p = 0.5)')
plt.xlabel('Number of Correct Guesses')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
# Confidence Interval for sample proportion
confidence_level = 0.95
z_critical = norm.ppf(1 - (1 - confidence_level) / 2)  # Z-score for 95% confidence
margin_of_error = z_critical * se
ci_lower = p_null - margin_of_error
ci_upper = p_null + margin_of_error
```

```python
# Question 8
# Findings and Discussion
# Conclusion regarding the Null Hypothesis
# If the p value is small we can regect the null hypothesis and assume the students are just guessing.
# If the p value is not small we cannot reject the null hypothesis so there is no proof that the students are not just guessing. 
```
